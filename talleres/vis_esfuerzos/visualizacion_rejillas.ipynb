# Visualización de una viga en voladizo con carga en el extremo libre

Se tiene una viga en voladizo con una fuerza por unidad de longitud en el extremo libre $F$, el extremo libre está ubicado en $x=0$. Los puntos $(x, y)$ satisfacen $x \in [0, L]$, $y \in [-h/2, h/2]$. Las componentes del tensor de esfuerzos están dadas por

\begin{align*}
\sigma_{xx} &= \frac{12 F}{h^3}xy\\
\sigma_{xy} &= \frac{3 F}{2 h} - \frac{6F}{h^3}y^2\\
\sigma_{yy} &= 0\, .
\end{align*}

```python
%matplotlib notebook
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 1
F = 1
L = 5
x = np.linspace(0, L, 201)
y = np.linspace(-h/2, h/2, 201)
y, x = np.meshgrid(y, x)
```

```python
Sxx = 12*F/h**3 * x*y
Sxy = 3*F/(2*h) - 6*F/h**3*y**2
Syy = np.zeros_like(Sxy)
```

```python
plt.figure()
plt.contourf(x, y, Sxx, cmap="RdBu_r")
```
